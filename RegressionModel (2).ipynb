```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
concrete_df=pd.read_csv('concrete_data.csv')
concrete_df.head()
```

```python
dcolumns=concrete_df.columns
predictors=concrete_df[dcolumns[dcolumns!='Strength']]
predictors.head()
```

```python
target=concrete_df['Strength']
target.head()
```

```python
n_cols=predictors.shape[1]
```

```python
#A
def regression_model():
    model=Sequential()
    model.add(Dense(10,activation='relu',input_shape=(n_cols,)))
    model.add(Dense(1))
    
    model.compile(optimizer='adam',loss='mean_squared_error')
    return model
```

```python
model=regression_model()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
lis=[]
for i in range(50):
    X_train,X_test,y_train,y_test=train_test_split(predictors,target,test_size=0.3,random_state=0)
    model.fit(X_train,y_train,validation_split=0.3,epochs=50,verbose=2)
    ypredict=model.predict(X_test)
    lis.append(mean_squared_error(ypredict,y_test))
```

```python
import statistics
from statistics import mean,pstdev
print(mean(lis),pstdev(lis))
```

```python
#B
predictors_norm=(predictors-predictors.mean())/predictors.std()
n_cols=predictors_norm.shape[1]
def regression_model():
    model=Sequential()
    model.add(Dense(10,activation='relu',input_shape=(n_cols,)))
    model.add(Dense(1))
    
    model.compile(optimizer='adam',loss='mean_squared_error')
    return model
```

```python
model=regression_model()
```

```python
lis1=[]
for i in range(50):
    X_train,X_test,y_train,y_test=train_test_split(predictors_norm,target,test_size=0.3,random_state=0)
    model.fit(X_train,y_train,validation_split=0.3,epochs=50,verbose=2)
    ypredict=model.predict(X_test)
    lis1.append(mean_squared_error(ypredict,y_test))
```

```python
print(mean(lis1),pstdev(lis1))
print('The mean of the mean squared error of B is greater than A')
```

```python
#D
predictors_norm=(predictors-predictors.mean())/predictors.std()
n_cols=predictors_norm.shape[1]
def regression_model():
    model=Sequential()
    model.add(Dense(10,activation='relu',input_shape=(n_cols,)))
    model.add(Dense(10,activation='relu'))
    model.add(Dense(10,activation='relu'))
    model.add(Dense(1))
    
    model.compile(optimizer='adam',loss='mean_squared_error')
    return model
```

```python
model=regression_model()
```

```python
lis3=[]
for i in range(50):
    X_train,X_test,y_train,y_test=train_test_split(predictors_norm,target,test_size=0.3,random_state=0)
    model.fit(X_train,y_train,validation_split=0.3,epochs=50,verbose=2)
    ypredict=model.predict(X_test)
    lis3.append(mean_squared_error(ypredict,y_test))
```

```python
print(mean(lis3),pstdev(lis3))
print('The mean of the mean squared error of D is lesser than B')
```
